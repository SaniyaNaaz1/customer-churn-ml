```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("../data/customer_churn_dataset-training-master.csv")

df.head()
```

```python
df.shape
df.info()
df.describe()
df.isnull().sum()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Churn", data=df)
plt.title("Churn Distribution")
plt.savefig("../assets/churn_distribution.png")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.savefig("../assets/correlation_heatmap.png")
plt.show()
```

```python
print(df.columns)
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Churn", y="Tenure", data=df)
plt.title("Tenure vs Churn")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Churn", y="Total Spend", data=df)
plt.title("Monthly Charges vs Churn")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Contract Length", hue="Churn", data=df)
plt.title("Contract Type vs Churn")
plt.xticks(rotation=45)
plt.show()
```

Customers with month-to-month contracts show significantly higher churn rates compared to yearly contracts. This suggests contract duration impacts retention.
